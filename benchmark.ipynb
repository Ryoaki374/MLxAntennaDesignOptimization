```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from deap import benchmarks
```

```python
x = np.arange(-32.768, 32.768, 0.5)
y = np.arange(-32.768, 32.768, 0.5)
X, Y = np.meshgrid(x, y)

Z = np.zeros(X.shape)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = benchmarks.ackley([X[i, j], Y[i, j]])[0]


## 3d plotting
fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(111, projection='3d')

ax.xaxis.pane.set_edgecolor('k')
ax.yaxis.pane.set_edgecolor('k')
ax.zaxis.pane.set_edgecolor('k')
ax.xaxis.pane.set_facecolor("w")
ax.yaxis.pane.set_facecolor("w")
ax.zaxis.pane.set_facecolor("w")

## surface plot
ax.plot_surface(X, Y, Z, cmap='turbo', edgecolor='none')

ax.set_title("DEAP Ackley Benchmark Function")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

#ax.grid(False)
ax.view_init(azim=35,elev=25)
ax.set_box_aspect((1,1,1))

plt.show()
```
